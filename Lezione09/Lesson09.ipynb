```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
```

```python
# Funzione per leggere i punti da file
def leggi_punti_da_file(nome_file):
    punti = []
    with open(nome_file, 'r') as file:
        for riga in file:
            x_str, y_str = riga.strip().split()
            x, y = float(x_str), float(y_str)
            punti.append((x, y))
    return punti
```

```python
nome_file = 'OUTPUT/results_cities_circle'

punti = leggi_punti_da_file(nome_file)
x_vals = [p[0] for p in punti]
y_vals = [p[1] for p in punti]

plt.plot(x_vals, y_vals, linestyle='-', marker='x', color='blue')
plt.title("Percorso migliore alla fine dell'ultima generazione \n - Cerchio -")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()
```

```python
x_vals = []
y_vals = []

with open('OUTPUT/results_genetic_circle', 'r') as file:
    for line in file:
        line = line.strip()
        
        x, y = map(float, line.split())
        
        x_vals.append(x)
        y_vals.append(y)

plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, marker = '.', label="Dati")
plt.xlabel("# Generazione")
plt.ylabel("<L^2>")
plt.title("Media della lunghezza del percorso sulla migliore metà della popolazione per generazione \n - Cerchio -")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
