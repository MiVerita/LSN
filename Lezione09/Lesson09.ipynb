```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
```

```python
# Funzione per leggere i punti da file
def leggi_punti_da_file(nome_file):
    punti = []
    with open(nome_file, 'r') as file:
        for riga in file:
            x_str, y_str = riga.strip().split()
            x, y = float(x_str), float(y_str)
            punti.append((x, y))
    return punti
```

```python
nome_file = 'OUTPUT/results_cities_circle'

punti = leggi_punti_da_file(nome_file)
x_vals = [p[0] for p in punti]
y_vals = [p[1] for p in punti]

plt.plot(x_vals, y_vals, linestyle='-', marker='x', color='blue')
plt.title("Percorso migliore alla fine dell'ultima generazione \n - Cerchio -")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()
```

### Miglior percorso trovato
In entrambi i  $TSP$ il miglior percorso trovato presenta tratti che non si intersecano fra loro, costituendo dunque la soluzione matematicamente ottimale del problema. La forma ordinata del percorso e l'assenza di "zig-zag" o salti bruschi testimonia la convergenza verso il miglior percorso possibile.

### Andamento di $L^2(x)$ del miglior individuo
In entrambi i $TSP$ si osserva una discesa esponenziale di $L^2(x)$ durante le prime generazioni $200 - 500$ seguita da una fase di stallo con variazioni limitate di $L^2(x)$ . Questo comportamento indica che l'algoritmo tende velocemente all'intorno del minimo di $L^2(x)$, ma rischia di ristagnare in soluzioni sub-ottimali per molte generazioni. Questo effetto si pu risolvere accentuando la diversit, ovvero aumentando le probabilit di crossover e di mutazioni. Cos facendo si stimola una maggiore esplorazione dello spazio delle soluzioni e favorendo l'uscita da minimi locali.

### Andamento di $\langle L^2(x) \rangle$ in funzione della generazione
$\langle L^2(x) \rangle$  calcolato sulla met migliore della popolazione e rappresenta un indicatore della fitness media degli individui. Anche in questo caso si osserva una forte decrescita per le prime $200 - 500$ generazioni. Successivamente, durante la fase di stagnazione di $L^2(x)$ del miglior individuo, $\langle L^2(x) \rangle$  presenta minime fluttuazioni. La forma piatta potrebbe essere sintomo di poca variabilit nella prima met della popolazione, con molte copie uguali. Per problemi pi complessi pu portare a non uscire da minimi locali.

### Commenti finali
In generale, anche cambiando la posizione delle citt nei due $TSP$, si osserva una convergenza verso il valore minimo della lunghezza del percorso dopo circa $200 - 500$ generazioni, confermando l'efficienza dell'algoritmo genetico nella risoluzione di $TSP$ di media complessit, con performance migliori al diminuire della difficolt del problema. In particolare, si nota una convergenza pi rapida verso la soluzione ottimale nel caso di citt disposte su una circonferenza, rispetto al caso in cui le citt sono distribuite casualmente allinterno di un quadrato. Questo  probabilmente dovuto alla presenza di simmetrie, che ne riduce la complessit.

E' importante scegliere attentamente le probabilit di crossover e mutazioni. In questo caso la probabilit di crossover  pari al $50$% mentre quella di mutazioni  pari al $10$%. Un valore elevato della probabilit di mutazione favorisce lesplorazione dello spazio delle soluzioni e pu aiutare a sfuggire da minimi locali, ma tende a rallentare la convergenza verso lottimo globale. Al contrario, un valore troppo basso limita lesplorazione, aumentando il rischio di convergenza prematura e stagnazione in soluzioni sub-ottimali.  quindi fondamentale trovare un equilibrio tra esplorazione (mutazioni) e selezione delle soluzioni migliori, adattando i parametri in base al problema.
